# Plotting in Python

To create an ordinary graph in Python we use the function plot from the pylab package. In the 
simplest case, this function takes one argument, which is a list or array of the values we want to plot. The 
function creates a graph of the given values in the memory of the computer, but it doesnâ€™t actually display it 
on the screen of the computer. The data are stored in the memory but not yet visible to the computer user. To 
display the graph we use a the plot and show functions within matplotlib, which take the graph in memory 
and draw it on the screen. Here is a complete program for plotting a small graph: 


```python
# 1-DIM PLOT

import matplotlib.pyplot as plt
y = [ 1.0, 2.4, 1.7, 0.3, 0.6, 1.8 ]
plt.plot(y)
plt.show()
```

```python
# 2-DIM PLOT

import matplotlib.pyplot as plt
x = [ 0.5, 1.0, 2.0, 4.0, 7.0, 10.0 ]
y = [ 1.0, 2.4, 1.7, 0.3, 0.6, 1.8 ]
plt.plot(x,y)
plt.show()
```

# SINE PLOT

Let us apply the plot and show functions to the creation of a slightly more interesting graph, a graph of the 
function sin x from x = 0 to x = 10. To do this we first create an array of the x values, then we take the sines of 
those values to get the y-coordinates of the points: 

```python
import matplotlib.pyplot as plt
from numpy import linspace,sin

x = linspace(0,10,100)           # generate the array of x-values,
y = sin(x)

plt.plot(x,y)
plt.show()
```

### (c) We have the temperatures and magnitudes of a set of stars in a file called stars.txt on our computer. This file is available on Canvas. Download it and upload it to your current working directory on the Jupyterhub. Within the file, the first column is the temperature and the second is the magnitude. Write a Python program to make a scatter plot of these data. Within your program, you should explore labeling the axes of the plot with the functions xlabel and ylabel, e.g. 

```python
import pandas as pd

# Read Text Files with Pandas using read_csv()

# read text file into pandas DataFrame and create
# header with x = frequency and y = Voltage

data = pd.read_csv("stars.txt", sep=" ", header=None,              
                names=["Temperature", "Magnitude"])

# display Data as a DataFrame
print(data)
```

```python
font1 = {'family':'serif','color':'blue','size':15}           # set font style
font2 = {'family':'serif','color':'darkred','size':15}        # set font style

plt.title("Number of Stars Data", fontdict = font1)           # title
plt.xlabel("Temperature", fontdict = font2)                   # x label
plt.ylabel("Magnitude", fontdict = font2)                     # y label

plt.scatter(x=data['Temperature'], y=data['Magnitude'])       #scatter command
plt.show()                                                    # show command
```
